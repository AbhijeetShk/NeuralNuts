```python
import torch
import torch.nn.functional as F


words = open('names.txt', 'r').read().splitlines()


chars = sorted(list(set(''.join(words))))
stoi = {s:i+1 for i,s in enumerate(chars)}
stoi['.'] = 0
itos = {i:s for s,i in stoi.items()}
vocab_size = len(stoi)   # 27
```

```python
xs1, xs2, ys = [], [], []

for w in words:
    chs = ['.', '.'] + list(w) + ['.']
    for ch1, ch2, ch3 in zip(chs, chs[1:], chs[2:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        iy  = stoi[ch3]

        xs1.append(ix1)
        xs2.append(ix2)
        ys.append(iy)

xs1 = torch.tensor(xs1)
xs2 = torch.tensor(xs2)
ys  = torch.tensor(ys)

num = xs1.nelement()
print("examples:", num)
```

```python
N = torch.zeros((vocab_size, vocab_size, vocab_size), dtype=torch.int32)

for a,b,c in zip(xs1, xs2, ys):
    N[a,b,c] += 1
```

```python
#smoothing

P = (N + 1).float()
P /= P.sum(2, keepdim=True)
P
```

```python
#loss eval

log_likelihood = 0.0
n = 0

for a,b,c in zip(xs1, xs2, ys):
    prob = P[a,b,c]
    log_likelihood += torch.log(prob)
    n += 1

nll = -log_likelihood
loss = nll / n

print("loss:", loss.item())
```

```python
g = torch.Generator().manual_seed(2147483647)

for i in range(20):

    out = []
    ix1 = 0
    ix2 = 0

    while True:

        p = P[ix1, ix2]
        ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()

        if ix == 0:
            break

        out.append(itos[ix])

        ix1 = ix2
        ix2 = ix

    print(''.join(out))
```
